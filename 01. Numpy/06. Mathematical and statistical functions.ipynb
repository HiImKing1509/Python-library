# **What is Numpy?**
**NumPy** là thư viện cơ bản cho tính toán khoa học bằng Python. Nó là một thư viện Python cung cấp một đối tượng mảng đa chiều, các đối tượng dẫn xuất khác nhau (chẳng hạn như các mảng và ma trận có mặt nạ) và một loạt các quy trình cho các hoạt động nhanh trên mảng, bao gồm toán học, logic, thao tác hình dạng, sắp xếp, lựa chọn, I / O , các phép biến đổi Fourier rời rạc, đại số tuyến tính cơ bản, các phép toán thống kê cơ bản, mô phỏng ngẫu nhiên và hơn thế nữa.

# **Mathematical and statistical functions**
Nhiều hàm toán học và thống kê có sẵn cho **ndarrays**

### **ndarrays** method
Một số hàm đơn giản là các phương thức **ndarray**

```python
a = np.array([[2, 5, 8], [-5, 7, 15], [10, 10.5, 6.2]])
a
```

```python
print("Mean =", a.mean())
```

```python
print("Min =", a.min())
```

```python
print("Max =", a.max())
```

Phương thức **prod()** trả về tích của mảng các phần tử trên một trục nhất định. Theo mặc định, trục được đặt thành Không, do đó tính toán tích của tất cả các phần tử trong mảng đã cho

```python
print("Prod =", a.prod())
```

Phương thức **std()**, được sử dụng để tính độ lệch chuẩn dọc theo trục được chỉ định. Hàm này trả về độ lệch chuẩn của các phần tử mảng. 

```python
print("Std =", a.std())
```

Theo mặc định, hàm **var()** tính toán phương sai tổng thể

```python
print("Var =", a.var())
```

Các hàm này chấp nhận một trục đối số tùy chọn cho phép thao tác được thực hiện trên các phần tử dọc theo trục đã cho

* Ví dụ: Hàm **sum()**

```python
b = np.arange(30).reshape(2, 3, 5)
b
```

**axis = 0**: Tổng các ma trận

```python
b.sum(axis = 0)
```

**axis = 1**: Tổng mỗi cột của mỗi ma trận tạo thành 1 **element**, mỗi ma trận là 1 **array**

```python
b.sum(axis = 1)
```

**axis = 2**: Tổng mỗi dòng của mỗi ma trận tạo thành 1 **element**, mỗi ma trận là 1 **array**

```python
b.sum(axis = 2)
```

### **Universal functions**
**Universal functions** là các trình bao bọc được vectơ hóa các chức năng đơn giản

```python
c = np.random.randn(2, 3, 4)
c
```

**abs()**: Trả về mảng cùng kích thước với ***trị tuyệt đối*** các phần tử

```python
np.abs(c)
```

**sqrt()**: Trả về mảng cùng kích thước với ***căn bậc 2*** các phần tử

```python
np.sqrt(c)
```

**log()**: Trả về mảng cùng kích thước với ***logarit cơ số e (logarit tự nhiên)*** các phần tử

```python
np.log(c)
```

**exp()**: Trả về mảng cùng kích thước với ***e mũ các phần tử***

```python
np.exp(c)
```

**sign()**: Trả về mảng đánh dấu cùng kích thước với

* Phần tử dương trả về 1
* Phần tử âm trả về -1
* Phần tử 0 trả về 0


```python
np.sign(c)
```

**ceil()**: Trả về mảng toàn giá trị nguyên với phần tử là ***phần tử nhỏ nhất lớn hơn hoặc bằng*** phần tử mảng ban đầu

```python
np.ceil(c)
```

**modf()**: ***Tách phần nguyên và phần thập phân riêng lẻ***, mỗi phần là một **array** cùng kích thước

```python
np.modf(c)
```

**isnan()**: Kiểm tra mảng có phần tử NaN hay không

```python
np.isnan(c)
```

**sin()**: Tính **sin** các phần tử trong mảng, tương tự với **cos**, **tan**, **arctan**

```python
print("sin =\n", np.sin(c))
print("cos =\n", np.cos(c))
print("tan =\n", np.tan(c))
print("cotan =\n", np.arctan(c))
```

### **Binary ufuncs**
Có nhiều hàm nhị phân, áp dụng theo từng phần tử trên hai **ndarrays**. Quy tắc **Broadcasting** được áp dụng nếu các mảng không có cùng hình dạng

```python
arr1 = np.array([1, -5, 7, -9])
arr2 = np.array([1.5, -2, -7. , 11])
```

**add(arr1, arr2)**: Tương đương với việc cộng hai mảng lại với nhau

```python
np.add(arr1, arr2)
```

**greater(arr1, arr2)**: So sánh các phần tử mảng **arr1** với các phần tử tương ứng mảng **arr2**, lớn hơn trả về **true** và ngược lại trả về **false**

```python
np.greater(arr1, arr2)
```

**maximum(arr1, arr2)**: Trả về **phần tử lớn nhất** giữa 2 phần tử tương ứng trong 2 mảng

```python
np.maximum(arr1, arr2)
```

**copysign(arr1, arr2)**: Giữ nguyên giá trị và đổi dấu phần tử mảng arr1 theo mảng arr2

```python
np.copysign(arr2, arr1)
```
