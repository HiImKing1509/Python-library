# **Categories**
> Th∆∞·ªùng xuy√™n xu·∫•t hi·ªán c√°c gi√° tr·ªã ƒë·∫°i di·ªán cho danh m·ª•c, v√≠ d·ª• *1* l√† n·ªØ, *2* l√† nam, ho·∫∑c *A* l√† t·ªët, *B* l√† trung b√¨nh, *C* l√† k√©m. **Pandas** c√≥ th·ªÉ x·ª≠ l√Ω nh·ªØng gi√° tr·ªã ƒë√≥ d·ªÖ d√†ng

```python
city_ = pd.DataFrame([["CA", "San Francisco", 37.781334, -122.416728],
                      ["NY", "New York", 40.705649, -74.008344],
                      ["FL", "Miami", 25.791100, -80.320733],
                      ["OH", "Cleveland", 41.473508, -81.739791],
                      ["UT", "Salt Lake City", 40.755851, -111.896657]],
                      columns=["state", "city", "lat", "lng"])
city_
```

```python
city_["eco_code"] = [12, 14, 14, 18, 12]
city_
```

T·∫°o m·ªôt c·ªôt ph√¢n lo·∫°i c√°c m√£ $ecocode$ tr√™n

```python
city_["economy"] = city_["eco_code"].astype('category')
city_["economy"].cat.categories
```
