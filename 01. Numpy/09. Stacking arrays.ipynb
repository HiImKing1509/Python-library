# **What is Numpy?**
**NumPy** là thư viện cơ bản cho tính toán khoa học bằng Python. Nó là một thư viện Python cung cấp một đối tượng mảng đa chiều, các đối tượng dẫn xuất khác nhau (chẳng hạn như các mảng và ma trận có mặt nạ) và một loạt các quy trình cho các hoạt động nhanh trên mảng, bao gồm toán học, logic, thao tác hình dạng, sắp xếp, lựa chọn, I / O , các phép biến đổi Fourier rời rạc, đại số tuyến tính cơ bản, các phép toán thống kê cơ bản, mô phỏng ngẫu nhiên và hơn thế nữa.

# **Numpy: Import**
Import thư viện **numpy**. Hầu hết mọi người sử dụng nó dưới dạng **np**

```python
import numpy as np
```

# **Stacking arrays**
Điều này thường có ích khi xếp chồng các mảng khác nhau lại với nhau

```python
arr1 = np.ones((3, 4))
arr2 = np.full((4, 4), 2.0)
arr3 = np.full((3, 4), 3.0)
```

```python
arr1
```

```python
arr2
```

```python
arr3
```

### **vstack**
Chồng các **arrays** lên với nhau, tất nhiên các **arrays** phải cùng chỉ số **column**

```python
arr4 = np.vstack((arr1, arr2, arr3))
arr4
```

```python
arr4.shape
```

### **hstack**
Sắp các **arrays** ngang hàng với nhau, tất nhiên các **arrays** phải cùng chỉ số **row**

```python
arr5 = np.hstack((arr1, arr3))
arr5
```

```python
arr5.shape
```

### **concatenate**
Nối các **arrays** với nhau theo bất kì trục nào dựa vào thuộc tính **axis**
* axis = 0: Chồng các array lên nhau (theo cột)
* axis = 1: Nối các array với nhau hàng ngang (theo dòng)

```python
arr6 = np.concatenate((arr1, arr2, arr3), axis = 0) # Tương tự như vstack
arr6
```

```python
arr7 = np.concatenate((arr1, arr3), axis = 1) # Tương tự như hstack
arr7
```

### **stack**
Hàm **stack** sắp xếp các **arrays** dọc theo một trục mới. Tất cả các **arrays** phải có cùng hình dạng

```python
arr8 = np.stack((arr1, arr3))
arr8
```

```python
arr8.shape
```
