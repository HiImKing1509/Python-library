### **Time range**
T·∫°o m·ªôt chu·ªói th·ªùi gian s·ª≠ d·ª•ng **pd.date_range()**. 

```python
dates = pd.date_range('16/11/2021 10:20am', periods=24, freq='H')
dates
```

Chu·ªói th·ªùi gian c√≥ th·ªÉ l√†m ch·ªâ m·ª•c trong **Series**

```python
temperatures = [19, 23, 21, 22, 16, 22, 12, 14, 43, 54, 23, 12, 12,
                23, 4, 2, 34 ,5, 34, 1, 32, 22, 13, 10]
temperatures
```

```python
dates_series = pd.Series(temperatures, dates)
dates_series
```

Plot **Series** ƒë√£ t·∫°o tr√™n

```python
import matplotlib.pyplot as plt

dates_series.plot(kind='bar')
plt.grid(True)
plt.show()
```

# **Resampling**

```python
dates_series_resample = dates_series.resample('3H') #Th√™m ph∆∞∆°ng th·ª©c nh∆∞ Min(), Max(), Mean(), ...
dates_series_resample
```

```python
dates_series_resample = dates_series_resample.mean()
dates_series_resample
```
