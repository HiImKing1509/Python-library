# **What is Numpy?**
**NumPy** là thư viện cơ bản cho tính toán khoa học bằng Python. Nó là một thư viện Python cung cấp một đối tượng mảng đa chiều, các đối tượng dẫn xuất khác nhau (chẳng hạn như các mảng và ma trận có mặt nạ) và một loạt các quy trình cho các hoạt động nhanh trên mảng, bao gồm toán học, logic, thao tác hình dạng, sắp xếp, lựa chọn, I / O , các phép biến đổi Fourier rời rạc, đại số tuyến tính cơ bản, các phép toán thống kê cơ bản, mô phỏng ngẫu nhiên và hơn thế nữa.

# **Vectorization**
> Thay vì thực hiện các hoạt động trên các mục mảng riêng lẻ, từng mục một, mã của 
chúng ta có thể hiệu quả hơn nhiều nếu cố gắng bám vào các hoạt động của mảng. Điều này gọi là **vector hóa**

**Ví dụ**:

Tạo một mảng **768x1024** dựa trên công thức **$sin(xy/40,5)$**. Một tùy chọn không tốt sẽ là thực hiện phép toán trong python bằng cách sử dụng các vòng lặp lồng nhau:

```python
import math
data = np.empty((768, 1024))
for i in range (768):
    for j in range (1024):
        data[i, j] = math.sin(i*j/40.5)
```

```python
data
```

Cách dùng trên hoạt động nhưng nó rất kém hiệu quả vì các vòng lặp đang diễn ra trong python thuần túy. 

Tiến hành vectơ hóa thuật toán này. Đầu tiên, chúng ta sẽ sử dụng hàm **meshgrid** của **NumPy** để tạo ma trận tọa độ từ các vectơ tọa độ.

```python
x_coords = np.arange(0, 1024)  # [0, 1, 2, ..., 1023]
y_coords = np.arange(0, 768)   # [0, 1, 2, ..., 767]
X, Y = np.meshgrid(x_coords, y_coords)
```
