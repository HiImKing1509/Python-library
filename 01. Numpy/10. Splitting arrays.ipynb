# **What is Numpy?**
**NumPy** là thư viện cơ bản cho tính toán khoa học bằng Python. Nó là một thư viện Python cung cấp một đối tượng mảng đa chiều, các đối tượng dẫn xuất khác nhau (chẳng hạn như các mảng và ma trận có mặt nạ) và một loạt các quy trình cho các hoạt động nhanh trên mảng, bao gồm toán học, logic, thao tác hình dạng, sắp xếp, lựa chọn, I / O , các phép biến đổi Fourier rời rạc, đại số tuyến tính cơ bản, các phép toán thống kê cơ bản, mô phỏng ngẫu nhiên và hơn thế nữa.

# **Splitting arrays**
Chia mảng ban đầu thành các mảng con

```python
arr1 = np.arange(54).reshape(9, 6)
arr1
```

Chia thành các phần bằng nhau **theo chiều dọc**

Có 2 cách thực hiện:
1.   Dùng phương thức **vsplit**, tức là phân chia trực tiếp theo chiều dọc
2.   Dùng phương thức chung **split** và sử dụng thuộc tính **axis = 0**
**Lưu ý**: Số lượng dòng phải chia hết cho số lượng mảng con cần phân chia


```python
arr2, arr3, arr4 = np.split(arr1, 3, axis = 0) # ~ np.vsplit(arr1, 3)
```

```python
arr2
```

```python
arr3
```

```python
arr4
```

Chia thành các phần bằng nhau **theo chiều ngang**

Có 2 cách thực hiện:
1.   Dùng phương thức **hsplit**, tức là phân chia trực tiếp theo chiều ngang
2.   Dùng phương thức chung **split** và sử dụng thuộc tính **axis = 1**
**Lưu ý**: Số lượng cột phải chia hết cho số lượng mảng con cần phân chia


```python
arr5, arr6 = np.split(arr1, 2, axis = 1) # ~ np.hsplit(arr1, 2)
```

```python
arr5
```

```python
arr6
```
